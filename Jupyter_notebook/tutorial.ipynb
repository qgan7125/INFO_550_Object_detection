# This is a simple tutorial to run Yolov5 

```python
# Make sure you install all packages
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
# Use small and fastest yolo pretained model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

# Detect example image

```python
# The example image
img = 'https://ultralytics.com/images/zidane.jpg'
```

```python
results = model(img)
results.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(results.render()))
plt.show()
```

# Real time object detection

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
